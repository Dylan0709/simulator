```python
from sympy import *
init_printing()

x = Symbol('x')

equation = x**2 + 3*x + 1

equation
```

```python
solve(equation, x)
```

```python
plot(equation)
```

```python
# Solve a differential equation

x = Function('x')
t = Symbol('t')

ode = Eq(x(t).diff(t), x(t)*t)

display(ode)

solution = dsolve(ode, x(t))

display(solution)

particular_solution = dsolve(ode, x(t), ics={x(0):1})

display(particular_solution)

plot(particular_solution.rhs, (t, -2, 2))
```
